```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
input_file='data/co2.dat'

years = []
data = []
annuals = []
annual_fits = []

infile = open(input_file,"r")
while infile:
    line = infile.readline()
    if len(line)==0:
        break
    data_list = line.split(" ")
    years.append(float(data_list[0]))
    annuals.append(float(data_list[13]))
    annual_fits.append(float(data_list[14]))
    for item in data_list[1:12]:
        data.append(float(item))
infile.close()
```

## Simple Plot

```python
plt.plot(data)
```

## Removed Outliers

```python
data_without_outliers = list(filter(lambda x: x>0, data))
```

```python
plt.plot(data_without_outliers)
```

## Annual Fit

```python
annual_fits_wo = []
for item in annual_fits:
    if item > 0:
        annual_fits_wo.append(item)
```

```python
plt.plot(annual_fits_wo)
```

## Labeled

```python
data_wo = []
years_12 = []
# remove outliers and pair the years w the data
for idx,i in enumerate(data):
    if i>0:
        data_wo.append(i)
        years_12.append(years[idx//12]+(idx%12)/12)
```

```python
plt.plot(years_12, data_wo)
plt.title("CO2 Concentrations at Mauna Loa\nMonthly Fit")
plt.xlabel("Year")
plt.ylabel("CO2 (ppm)")
```

```python
annual_fits_wo = []
years_wo = []

# do the same thing as last plot
for i,j in zip(annual_fits,years):
    if i>0:
        annual_fits_wo.append(i)
        years_wo.append(j)
```

```python
plt.plot(years_wo,annual_fits_wo)
plt.title("CO2 Concentrations at Mauna Loa\nAnnual Fit")
plt.xlabel("Year")
plt.ylabel("CO2(ppm)")
```
