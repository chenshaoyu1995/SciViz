```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import random
import scipy
```

```python
## scaller plot
```

```python
num_samples = 100
variance = 5

x = []
y = []

delta = num_samples/100.0

for i in range(0,num_samples):
    x.append(i*delta+random.uniform(-variance,variance))
    y.append(i*delta+random.uniform(-variance,variance))
```

```python
fig = plt.figure()
plt.setp(fig, facecolor='w')

points = plt.plot(x,y,c='k',marker='o',linestyle='None')
# plt.setp(points,ms=7,markerfacecolor=[1,1,1,0],markeredgewidth=1.5)
plt.setp(points,ms=4)
plt.title('Correlation')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([-10,110,-10,110])
plt.xticks([0,50,100])
plt.yticks([0,50,100])
```

```python
## correlation
```

```python
(ar,br) = scipy.polyfit(x,y,1)
ry = scipy.polyval([ar,br],x)
rx = x

fig = plt.figure()
plt.setp(fig, facecolor='w')

points = plt.plot(x,y,c='k',marker='o',linestyle='None')
#plt.setp(points,ms=7,markerfacecolor=[1,1,1,0],markeredgewidth=1.5)
plt.setp(points,ms=4)
plt.corr = plt.plot(rx,ry,linewidth=3.0,c=[0.5,0.5,0.5])
#line = plot([-5,105],[-5,105],linewidth=3.0,c=[1.0, 0.0, 0.0], linestyle='--')
plt.title('Correlation')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([-10,110,-10,110])
plt.xticks([0,50,100])
plt.yticks([0,50,100])
#legend(['pseudo random data','least squares fit','perfect fit'],loc=2)
plt.legend(['pseudo random data','least squares fit'],loc=2)
```

```python
## Exact Fit Option
```

```python
on = True

(ar,br) = scipy.polyfit(x,y,1)
ry = scipy.polyval([ar,br],x)
rx = x

fig = plt.figure()
plt.setp(fig, facecolor='w')

points = plt.plot(x,y,c='k',marker='o',linestyle='None')
#setp(points,ms=7,markerfacecolor=[1,1,1,0],markeredgewidth=1.5)
plt.setp(points,ms=4)
corr = plt.plot(rx,ry,linewidth=3.0,c=[0.5,0.5,0.5])
if on:
    line = plt.plot([-5,105],[-5,105],linewidth=3.0,c=[1.0, 0.0, 0.0], linestyle='--')
plt.title('Correlation')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([-10,110,-10,110])
plt.xticks([0,50,100])
plt.yticks([0,50,100])
#legend(['psuedo random data','least squares fit','perfect fit'],loc=2)
if on:
    plt.legend(['pseudo random data','least squares fit','exact fit'],loc=2)
else:
    plt.legend(['pseudo random data','least squares fit'],loc=2)
```
