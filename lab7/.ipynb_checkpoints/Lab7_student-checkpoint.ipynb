## Comparing Euler and Runge-Kutta for Streamlines computation

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

n = 50

xlim = np.array([-5,5])
ylim = np.array([-5,5])
dxy = (xlim[1]-xlim[0])/n
X, Y = np.mgrid[xlim[0]:xlim[1]:dxy,ylim[0]:ylim[1]:dxy]
U, V = -Y, X
#U, V = Y-X, -X-Y

plt.quiver(X, Y, U, V, edgecolor='k', facecolor='None', linewidth=0.5)
```

```python
def bilinear_interpolation(x,y,u,v):
    f00 = np.array([u[0,0], v[0,0]])
    f10 = np.array([u[1,0], v[1,0]])
    f01 = np.array([u[0,1], v[0,1]])
    f11 = np.array([u[1,1], v[1,1]])
    return(np.multiply((1-x)*(1-y),f00) +
           np.multiply(x*(1-y),f10) +
           np.multiply(y*(1-x),f01) +
           np.multiply(x*y,f11))
```

```python
# Euler integration
nseeds = 10
x0 = np.zeros((nseeds,2))
x0[:,0] = np.random.uniform(low=xlim[0] + 2 * dxy, high=xlim[1] - 2 * dxy, size=nseeds)
x0[:,1] = np.random.uniform(low=ylim[0] + 2 * dxy, high=ylim[1] - 2 * dxy, size=nseeds)

nsteps = 1500
pos_euler = np.zeros((nseeds, nsteps, 2))
pos_euler[:,0] = x0
dt = 1e-2

for t in range(1,nsteps):
    for s in range(nseeds): 
        i,j = np.floor((pos_euler[s, t - 1] - np.array([xlim[0], ylim[0]])) / dxy).astype(int)
        if (i < 0) or (i > X.shape[0] - 2) or (j < 0) or (j > X.shape[1] - 2):
            pos_euler[s,t] = pos_euler[s,t-1]
        else:
            # Euler step
            uv_interp = bilinear_interpolation(pos_euler[s, t - 1, 0] - X[i, j], 
                                               pos_euler[s, t - 1, 1] - Y[i, j], 
                                               U[i:i + 2,j:j + 2], 
                                               V[i:i + 2,j:j + 2])
            pos_euler[s, t] = pos_euler[s, t - 1] + dt * uv_interp # Euler step

plt.quiver(X[::2], Y[::2], U[::2], V[::2], edgecolor='k', facecolor='None', linewidth=0.5)
for s in range(nseeds):
    plt.plot(pos_euler[s,:,0],pos_euler[s,:,1])
```
