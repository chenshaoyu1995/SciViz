```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Simple Plot

```python
mammals = []
sizes = []
rates = []

infile = open("data/scaling.dat","r")
while infile:
    line = infile.readline()
    if len(line)==0:
        break
    data_list = line.split(":")
    mammals.append(data_list[0])
    sizes.append(float(data_list[1]))
    rates.append(float(data_list[2]))
infile.close()
```

```python
fig = plt.figure()
plt.setp(fig, facecolor='w')

points = plt.scatter(sizes,rates,c='k',marker='o',s=30)
plt.title('Body Size and Metabolic Rate')
plt.xlabel('Body Weight (kg)')
plt.ylabel('Heat Production per Day (kcal)')
```

## Log Log Scale

```python
points = plt.loglog(sizes,rates,c='k',marker='o',linestyle='None')
plt.title('Body Size and Metabolic Rate')
plt.xlabel('Body Weight (kg)')
plt.ylabel('Heat Production per Day (kcal)')
plt.xticks([1e-2, 1, 100, 10000],['0.01', '1', '100', '10,000'])
plt.yticks([1, 100, 10000, 1000000],['1','100','10,000','1,000,000'])
```

## Labels

```python
points = plt.loglog(sizes,rates,c='k',marker='o',linestyle='None')
plt.title('Body Size and Metabolic Rate')
plt.xlabel('Body Weight (kg)')
plt.ylabel('Heat Production per Day (kcal)')
plt.xticks([1e-2, 1, 100, 10000],['0.01', '1', '100', '10,000'])
plt.yticks([1, 100, 10000, 1000000],['1','100','10,000','1,000,000'])

for i in range(0,len(sizes)):
    plt.text(sizes[i],rates[i]-rates[i]/2,mammals[i])
```

## Improves Labels

```python
fig = plt.figure()
plt.setp(fig, facecolor='w')
points1 = plt.loglog(sizes,rates,c='w',marker='o',linestyle='None')
points2 = plt.loglog(sizes,rates,c='k',marker='o',linestyle='None')
plt.setp(points1,ms=10,markerfacecolor=[1,1,1,0],markeredgewidth=1.5)
plt.setp(points2,ms=5)
plt.title('Body Size and Metabolic Rate')
plt.xlabel('Body Weight (kg)')
plt.ylabel('Heat Production per Day (kcal)')
plt.axis([1e-3,30000,1,2000000])
plt.xticks([1e-2, 1, 100, 10000],['0.01', '1', '100', '10,000'])
plt.yticks([1, 100, 10000, 1000000],['1','100','10,000','1,000,000'])

for i in [0,1,3,6,11,12,13,14]:
    plt.text(sizes[i]+sizes[i]/4,rates[i]-rates[i]/5,mammals[i])
#others
plt.text(sizes[2]-sizes[2]/1.8,rates[2]-rates[2]/5,mammals[2])
plt.text(sizes[5]-sizes[5]/4,rates[5]-rates[5]/2.7,mammals[5])
plt.text(sizes[4]-sizes[4]/1.3,rates[4]-rates[4]/6,mammals[4])
plt.text(sizes[7]-sizes[7]/1.6,rates[7]-rates[7]/4.5,mammals[7])
plt.text(sizes[8]-sizes[8]/1.6,rates[8]-rates[8]/2.5,mammals[8])
plt.text(sizes[9]-sizes[9]/1.12,rates[9]+rates[9]/10,mammals[9])
plt.text(sizes[10]-sizes[10]/4,rates[10]+rates[10]/4,mammals[10])
```
