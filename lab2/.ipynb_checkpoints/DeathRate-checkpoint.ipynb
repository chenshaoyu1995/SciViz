```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Default Plot

```python
causes = []
odds = []

infile = open("data/death.dat","r")
while infile:
    line = infile.readline()
    if len(line)==0:
        break
    data_list = line.split(":")
    causes.append(data_list[0])
    odds.append(1.0/float(data_list[1]))
infile.close()
```

```python
causes
```

```python
y = range(0,len(odds))
plt.plot(odds,y,'ko')
plt.title('Causes of Death over Lifetime')
plt.xlabel('Percentage')
plt.yticks(y,causes)
plt.xticks([0,0.05,0.1,0.15,0.2,0.25],['0%','5%','10%','15%','20%'])
```

## Reformatted

```python
fig = plt.figure()
plt.setp(fig, facecolor='w')

y = range(0,len(odds))
points = plt.plot(odds,y,'ko')
plt.setp(points,ms=8.0)
plt.title('Causes of Death over Lifetime')
plt.xlabel('Percentage')
plt.yticks(y,causes)
plt.xticks([0,0.1,0.2],['0%','10%','20%'])
plt.axis([-0.005, 0.21,-1,len(odds)])
```

## Flip Vertically

```python
flip_causes = []
flip_odds= []

for i in range(len(causes)):
    flip_causes.append(causes[-i-1])
    flip_odds.append(odds[-i-1])
```

```python
fig = plt.figure()
plt.setp(fig, facecolor='w')

y = range(0,len(flip_odds))
points = plt.plot(flip_odds,y,'ko')
plt.setp(points,ms=8.0)
plt.title('Causes of Death over Lifetime')
plt.xlabel('Percentage')
plt.yticks(y,flip_causes)
plt.xticks([0,0.1,0.2],['0%','10%','20%'])
plt.axis([-0.005, 0.21,-1,len(flip_odds)])
```

## Grid Lines

```python
fig = plt.figure()
plt.setp(fig, facecolor='w')

y = range(0,len(flip_odds))
points = plt.plot(flip_odds,y,'ko')
plt.setp(points,ms=8.0)
plt.title('Causes of Death over Lifetime')
plt.xlabel('Percentage')
plt.yticks(y,flip_causes)
plt.xticks([0,0.1,0.2],['0%','10%','20%'])
plt.axis([-0.005, 0.21,-1,len(flip_odds)])
plt.grid(True)
plt.gca().xaxis.grid(False)
```

## Log Scale

```python
fig = plt.figure()
plt.setp(fig, facecolor='w')

y = range(0,len(flip_odds))
points = plt.semilogx(odds,y,'ko')
plt.setp(points,ms=8.0)
plt.title('Causes of Death over Lifetime')
plt.xlabel('Percentage (Log10)')
plt.yticks(y,flip_causes)
plt.xticks([1.0e-5,1.0e-4,1.0e-3,1.0e-2,1.0e-1],['0.001%','0.01%','0.1%','1%','10%'])
plt.axis([0.2e-5, 0.3,-1,len(odds)])
plt.grid(True)
plt.gca().xaxis.grid(False)
```
