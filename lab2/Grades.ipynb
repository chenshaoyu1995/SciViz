```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Basic Box Plot

```python
grades = []

infile = open("data/grades1.dat","r")
while infile:
    line = infile.readline()
    if len(line)==0:
        break
    grades.append(float(line))
infile.close()

grades2 = []
infile = open("data/grades2.dat","r")
while infile:
    line = infile.readline()
    if len(line)==0:
        break
    grades2.append(float(line))
infile.close()
```

```python
fig = plt.figure()
plt.setp(fig, facecolor='w')

bp = plt.boxplot(grades)
plt.title('Midterm 2 Grades')
plt.ylabel('Percentage')
plt.axis([0.85,1.15,0,105])
```

## Improved Plot

```python
fig = plt.figure()
plt.setp(fig, facecolor='w')

bp = plt.boxplot(grades,sym='bo')
plt.title('Midterm 2 Grades')
plt.ylabel('Percentage')
plt.axis([0.85,1.15,30,105])
plt.setp(bp['medians'],linewidth=3.0,c='k')
plt.setp(bp['fliers'],linewidth=3.0,c='k')
plt.setp(bp['whiskers'],linewidth=3.0,c='k')
plt.setp(bp['boxes'],linewidth=3.0,c='k')
plt.setp(bp['caps'],linewidth=3.0,c='k')
```
